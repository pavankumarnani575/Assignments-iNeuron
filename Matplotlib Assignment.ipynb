 ### Scipy:
    We have the min and max temperatures in a city In India for each months of the year.
    We would like to find a function to describe this and show it graphically, the dataset
    given below.
    Task:
    1.fitting it to the periodic function
    2.plot the fit
    Data
    Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
    Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from scipy.optimize import curve_fit
```

```python
x=np.arange(0,12)
y_max=np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
y_min=np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
def func(x,a,b,c,d):
    return a+b*np.cos((x +c) * 1.8 * np.pi / x.max())
opt_max,cov_max=curve_fit(func,x,y_max)
opt_min,cov_min=curve_fit(func,x,y_min)
```

```python
x_fit=np.linspace(0,12)
```

```python
fig=plt.figure(figsize=(7,5))
ax=fig.add_axes([0,0,1,1])
ax.scatter(x,y_min,color='blue',label='Male',marker='o',s=50)
ax.scatter(x,y_max,color='red',label='Male',marker='o',s=50)
ax.set_xlim(0,12)
ax.set_xlabel('Month',fontsize=15,color='indigo')
ax.set_ylabel('Temperature ($^\circ$C)',fontsize=15,color='indigo')
ax.set_title('Temperature Details in a year',fontsize=20,color='blue')
ax.spines['bottom'].set_color('green')
ax.spines['left'].set_color('green')
ax.spines['bottom'].set_linewidth(3)
ax.spines['left'].set_linewidth(3)

ax.plot(x_fit,func(x_fit,*opt_max),color='red')
ax.plot(x_fit,func(x_fit,*opt_min),color='blue')
sb.set_style('white')
plt.show()
```

 ### Matplotlib:
    This assignment is for visualization using matplotlib:
    data to use:
    url=https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv
    titanic = pd.read_csv(url)
    Charts to plot:
    1. Create a pie chart presenting the male/female proportion
    2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
```

```python
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic=pd.read_csv(url)
```

```python
titanic['sex'].value_counts()
```

```python
male_fare=titanic[titanic['sex']=='male'][['age','fare']]
female_fare=titanic[titanic['sex']=='female'][['age','fare']]
```

```python
fig,ax=plt.subplots(1,2,figsize=(15,7))
ax[0].pie(titanic.sex.value_counts(),labels=['Male','Female'],autopct='%1.2f%%',shadow = True,explode=(0,0.1),colors=['red','green'])
ax[0].set_title('Male & Female Proportion',fontsize=20,color='blue')
ax[1].scatter(male_fare['age'],male_fare['fare'],color='red',label='Male',marker='x',s=50)
ax[1].scatter(female_fare['age'],female_fare['fare'],color='green',label='Female',marker='+',s=50)
ax[1].set_title('Age & Fare Plot for Male and Female',fontsize=20,color='blue')
ax[1].set_xlabel('Age',fontsize=15,color='indigo')
ax[1].set_ylabel('Fare',fontsize=15,color='indigo')
ax[1].legend(fontsize=14)
ax[1].spines['bottom'].set_color('blue')
ax[1].spines['left'].set_color('blue')
ax[1].spines['bottom'].set_linewidth(3)
ax[1].spines['left'].set_linewidth(3)
sb.set_style('dark')
fig.set(facecolor='lightgreen')
plt.show()
```
